```python
import torch
from matplotlib import pyplot as plt
import numpy as np
```

```python
GT=torch.load('Results/short_GT.pt')
DNN=torch.load('Results/short_DNN.pt')
GRU=torch.load('Results/short_GRU.pt')
TCN=torch.load('Results/short_TCN.pt')
EncoderOnly=torch.load('Results/short_EncoderOnly.pt')
DecoderOnly=torch.load('Results/short_DecoderOnly.pt')
V_Former=torch.load('Results/short_V_Former.pt')
G_Former=torch.load('Results/short_G_Former.pt')
```

```python

plt.rcParams['figure.figsize']=(7, 5)
plt.plot(GT,linewidth=2.5,label='GroudTruth',color='black')
plt.plot(DNN,label='DNN')
plt.plot(GRU,label='GRU')
plt.plot(TCN,label='TCN')
plt.plot(EncoderOnly,label='EncoderOnly')
plt.plot(DecoderOnly,label='DecoderOnly')
plt.plot(V_Former,label='V_Former')
plt.plot(G_Former,label='G_Former',color='red')
plt.legend(loc='lower right')
plt.xlabel('Time(15min)')
plt.savefig('ShortCompare.svg',dpi=1000,bbox_inches='tight')
```

```python
GT=torch.load('Results/long_GT.pt')
DNN=torch.load('Results/long_DNN.pt')
GRU=torch.load('Results/long_GRU.pt')
TCN=torch.load('Results/long_TCN.pt')
EncoderOnly=torch.load('Results/long_EncoderOnly.pt')
DecoderOnly=torch.load('Results/long_DecoderOnly.pt')
V_Former=torch.load('Results/long_V_Former.pt')
G_Former=torch.load('Results/long_G_Former.pt')
```
